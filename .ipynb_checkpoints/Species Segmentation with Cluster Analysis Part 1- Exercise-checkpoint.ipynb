There are 4 features: sepal length, sepal width, petal length, and petal width.

Start by creating 2 clusters. Then standardize the data and try again. Does it make a difference?

Use the Elbow rule to determine how many clusters are there.


## Import the relevant libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

# Fix the memory leak
import os
os.environ["OMP_NUM_THREADS"] = "1"

from sklearn.cluster import KMeans
from sklearn import preprocessing
```

## Load the data

Load data from the csv file: <i> 'iris_dataset.csv'</i>.

```python
raw_data = pd.read_csv('iris-dataset.csv')
data = raw_data.copy()
data.head()
```

## Plot the data

For this exercise, try to cluster the iris flowers by the shape of their sepal. 

<i> Use the 'sepal_length' and 'sepal_width' variables.</i> 

```python
x = data.iloc[:, 0:2]
x.head()
```

# Clustering (unscaled data)

Separate the original data into 2 clusters.

```python
kmeans = KMeans(2)
kmeans.fit(x)
```

```python
identified_clusters = kmeans.fit_predict(x)
data['clusters'] = identified_clusters
data.head()
```

```python
plt.scatter(data['sepal_length'], data['sepal_width'], c=data['clusters'], cmap='rainbow')
plt.xlabel('Sepal Length', size=16)
plt.ylabel('Sepal Width', size=16)
plt.title('Sepal Clusters')
plt.show()
```

# Standardize the variables

Import and use the <i> method </i> function from sklearn to standardize the data. 

```python
x_scaled = preprocessing.scale(x)
x_scaled
```

# Clustering (scaled data)

```python
kmeans = KMeans(3)
kmeans.fit(x_scaled)
```

```python
clusters_scaled = kmeans.fit_predict(x_scaled)
```

```python
data['standardized_scale'] = clusters_scaled
```

```python
plt.scatter(data['sepal_length'], data['sepal_width'], c=data['standardized_scale'], cmap='rainbow')
plt.xlabel('Sepal Length', size=16)
plt.ylabel('Sepal Width', size=16)
plt.title('Sepal Clusters')
plt.show()
```

## Take Advantage of the Elbow Method

### WCSS

```python
wcss = []

for i in range(1,9):
    kmeans = KMeans(i)
    kmeans.fit(x_scaled)
    wcss_iter = kmeans.inertia_
    wcss.append(wcss_iter)

wcss
```

### The Elbow Method

```python
n_clusters = range(1,9)
plt.plot(n_clusters, wcss)
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()

# possible cluster numbers are: 2, 3, 5
```
